# **Comparing Performance of various models from Model Zoo**

Now, we compare the performance of various models and report their performance.

We select bbox mean average precision (mAP) to compare the various models.
We select mAP because there are two labels - ball and player. To consider both the labels, we select mean average precision, and get the mean of average precisions of both the labels.

Also, average precision uses the precision-recall graph area for a given IoU value. It combines both precision and recall, since there is no particular preference for either precision and recall in our case (both false positives and false negatives are equally bad, one is not worse thant the other).
